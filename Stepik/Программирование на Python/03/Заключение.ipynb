Всем желающим можно обратить внимание на продолжение курса по Python:
1) Python: основы и применение - https://stepik.org/course/512. Будет рассказано про:
- классы, объекты, исключения, наследования, разные крутые штуки и конечно же работа с разными форматами данных. 
Будет сложновато, даже после этого курса. 

2) Чуть больше задачек по Python есть тут - https://stepik.org/course/431/.

Если у вас появилось желание писать программы лучше - то советуем курсы по алгоритмам:

3) Алгоритмы: теория и практика. Структуры данных - https://stepik.org/course/217/syllabus
Расскажет о том, какие структуры данных существуют и в каких случаях лучше применять одну структуру, а в каких - другую. Конечно же, всё придётся реализовывать самому.

4) Алгоритмы: теория и практика. Методы - https://stepik.org/course/1547/syllabus
Расскажет о том, что такое сложность алгоритма, выды алгоритмов, их различия и плюсы. Вам нужно будет поделать теоретические задачки (но уже без оценок), реализовать каждый из алгоритмов. Легко не будет, но после этого любая сложная задача перестанет вызывать ужас, вы будете применять подход "Разделяй и властвуй", после чего любая задача вам поддастся.

5) Введение в Linux - https://stepik.org/course/73 
Познакомит вас с Линуксом и его консолью. Чтоб на MacBook можно было работать как профессионал )))

6) Введение в архитектуру ЭВМ. Элементы операционных систем. - https://stepik.org/course/253/syllabus
Поможет разобраться и понять, как же работает железо в компьютерах.
